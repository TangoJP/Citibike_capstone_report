# Predicting the Daily Ridership

Same as the other regression notebook except that 10-day, 30-day, and 90-day prediction windows were tried here and compared

```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
%matplotlib inline
```

```python
# Scoring function for a regressor
def score_regressor(regressor, X, y, X_train=None, y_train=None):
    
    result = {}
    
    y_pred = regressor.predict(X)
    
    # calculate R^2 and MSE
    r2 = r2_score(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    
    scores = {'r2_score': r2,
                     'mse': mse}
    
    result['scores'] = scores
    
    # If train option is on, return results for the training set too
    if (X_train is not None) & (y_train is not None):
        
        y_pred_train = regressor.predict(X_train)
        
        r2_train = r2_score(y_train, y_pred_train)
        mse_train = mean_squared_error(y_train, y_pred_train)

        scores_train = {'r2_score': r2_train,
                                    'mse': mse_train}
        
        result['scores_train'] = scores_train
    
    # If the regressor has feature importances attribute, include in the result
    try:
        result['feature_importances'] = regressor.feature_importances_
    except:
        pass
    
    return result

# A method to print out the regression scores
def print_regressor_scores(result, trainset=False, feature_labels=None):
    
    print('==== Regressor Performance ====')
    if trainset:
        print('Train: R2 = %f, mse = %f'  % (result['scores_train']['r2_score'], result['scores_train']['mse']))
    print('Test: R2 = %f, mse = %f'  % (result['scores']['r2_score'], result['scores']['mse']))
    
    if (feature_labels is not None):
        importances = result['feature_importances']
        indices = np.argsort(importances)[::-1]
        print('--------------- Feature Importances ---------------')
        for f in range(len(feature_labels)):
            print('%2d) %-*s %f' % (f + 1, 20, feature_labels[indices[f]], importances[indices[f]]))
        print('\n')
        
    return
```

# Predicting future usage by walk-forward testing
- Decide on a cutoff date at which to split the dataset into train/test sets 
- Fit model with train set, test its perfromance on test set
- Shift cutoff date by certain interval, repeat the train/test cycle
- Assess how performance change over time and identify optimal cutoff date

<b>Predicting only a certain period following the cutoff date</b>
- Test set is the remaining data for the 10, 30, and 90days following the cutoff date

```python
%%time
# Set up initial split date, end date for iteration, and shift interval
# Now window is also set up as a prediction window
split_date = dt.date(2014, 1, 1)
end_date = dt.date(2016, 12, 31)
shift = dt.timedelta(days=10)
window10 = dt.timedelta(days=10)
window30 = dt.timedelta(days=30)
window90 = dt.timedelta(days=90)

# Set up results container
split_dates = []
train_r2s = []
test_r2s_10 = []
test_r2s_30 = []
test_r2s_90 = []
train_mses = []
test_mses_10 = []
test_mses_30 = []
test_mses_90 = []
counter = []

# Iterate until the iteration reaches the end_date
i = 0
while split_date <= end_date:
    
    # Split the data
    past = data[data['date'] <= split_date]
    future10 = data[(data['date'] > split_date) & (data['date'] <= (split_date + window10))]
    future30 = data[(data['date'] > split_date) & (data['date'] <= (split_date + window30))]
    future90 = data[(data['date'] > split_date) & (data['date'] <= (split_date + window90))]
    
    # Set up Xs and ys for the past
    X_past = past[past.columns[1:-1]]
    y_past = past[past.columns[-1]]
    X_past = np.array(X_past).reshape(X_past.shape)
    y_past = np.array(y_past).reshape(-1, 1).ravel()
    
    # Set up Xs and ys for the past for the futures
    X_future10 = future10[future10.columns[1:-1]]
    y_future10 = future10[future10.columns[-1]]
    X_future10 = np.array(X_future10).reshape(X_future10.shape)
    y_future10 = np.array(y_future10).reshape(-1, 1).ravel()
    
    X_future30 = future30[future30.columns[1:-1]]
    y_future30 = future30[future30.columns[-1]]
    X_future30 = np.array(X_future30).reshape(X_future30.shape)
    y_future30 = np.array(y_future30).reshape(-1, 1).ravel()
    
    X_future90 = future90[future90.columns[1:-1]]
    y_future90 = future90[future90.columns[-1]]
    X_future90 = np.array(X_future90).reshape(X_future90.shape)
    y_future90 = np.array(y_future90).reshape(-1, 1).ravel()
    
    # Train RandomForestRegressor
    regressor = RandomForestRegressor(max_depth=40, n_estimators=50, n_jobs=4)
    regressor.fit(X_past, y_past)
    
    # Score with the test set (future set)
    result10 = score_regressor(regressor, X_future10, y_future10, X_train=X_past, y_train=y_past)
    result30 = score_regressor(regressor, X_future30, y_future30)
    result90 = score_regressor(regressor, X_future90, y_future90)
    
    train_r2s.append(result10['scores_train']['r2_score'])
    train_mses.append(result10['scores_train']['mse'])
    
    test_r2s_10.append(result10['scores']['r2_score'])
    test_mses_10.append(result10['scores']['mse'])
    
    test_r2s_30.append(result30['scores']['r2_score'])
    test_mses_30.append(result30['scores']['mse'])
    
    test_r2s_90.append(result90['scores']['r2_score'])
    test_mses_90.append(result90['scores']['mse'])
    
    # print progress
    if (i % 9) == 0:
        print('Split at', split_date, ': R^2s = %f, %f, %f' \
                                      % (result10['scores']['r2_score'], result30['scores']['r2_score'], result90['scores']['r2_score']))
    
    # Store split date and the counter
    split_dates.append(split_date)
    counter.append(i)
    
    # update the split date and the counter
    i += 1
    split_date += shift

print('...Process Complete.')
```

```python
# set up dates for drawing vertical lines to indicate beginnings of the year
vline1 = dt.date(2014, 1, 1)
vline2 = dt.date(2015, 1, 1)
vline3 = dt.date(2016, 1, 1)
vline4 = dt.date(2017, 1, 1)
```

```python
# Plot the results in separate panels
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(20, 14))

axes = [(ax1, ax4), (ax2, ax5), (ax3, ax6)]
results = [(test_r2s_10, test_mses_10), (test_r2s_30, test_mses_30), (test_r2s_90, test_mses_90)]
labels = ['10-day window', '30-day window', '90-day window']
colors = ['turquoise', 'salmon', 'goldenrod']

for i, axpair in enumerate(axes):
    axpair[0].plot(split_dates, train_r2s, label='Train')
    axpair[0].plot(split_dates, results[i][0], c=colors[i], label='%s' % labels[i])
    #axpair[0].set_xlabel('Data Split Date', fontsize=14)
    axpair[0].set_ylabel('R^2 score', fontsize=14)
    axpair[0].set_title('R^2 score: %s\n ' % labels[i], fontsize=16)
    axpair[0].legend(loc='lower right', fontsize=14)
    axpair[0].set_ylim([-0.05, 1.05])
    axpair[0].axvline(vline1, ls='--', c='0.8')
    axpair[0].axvline(vline2, ls='--', c='0.8')
    axpair[0].axvline(vline3, ls='--', c='0.8')
    axpair[0].axvline(vline4, ls='--', c='0.8')
    
    axpair[1].plot(split_dates, train_mses, label='Train')
    axpair[1].plot(split_dates,results[i][1], c=colors[i], label='%s' % labels[i])
    axpair[1].set_xlabel('Data Split Date', fontsize=14)
    axpair[1].set_ylabel('Mean Squared Error (MSE)', fontsize=14)
    axpair[1].set_title('MSE: %s\n ' % labels[i], fontsize=16)
    axpair[1].legend(loc='upper right', fontsize=14)
    axpair[1].set_ylim([-50, 2050])
    axpair[1].axvline(vline1, ls='--', c='0.8')
    axpair[1].axvline(vline2, ls='--', c='0.8')
    axpair[1].axvline(vline3, ls='--', c='0.8')
    axpair[1].axvline(vline4, ls='--', c='0.8')

    for tick in axpair[0].get_xticklabels():
            tick.set_rotation(45)
    for tick in axpair[1].get_xticklabels():
            tick.set_rotation(45)

plt.tight_layout()
```

```python
# print out the best R^2 score and the split date for it
best_score10 = np.max(test_r2s_10)
best_score30 = np.max(test_r2s_30)
best_score90 = np.max(test_r2s_90)
best_split_date10 = split_dates[np.argmax(test_r2s_10)]
best_split_date30 = split_dates[np.argmax(test_r2s_30)]
best_split_date90 = split_dates[np.argmax(test_r2s_90)]
print('Best R^2 score for 10-day window = %f on %s' % (best_score10, str(best_split_date10)))
print('Best R^2 score for 30-day window = %f on %s' % (best_score30, str(best_split_date30)))
print('Best R^2 score for 90-day window = %f on %s' % (best_score90, str(best_split_date90)))
```
